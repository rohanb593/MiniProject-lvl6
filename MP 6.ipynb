### whileLoop

#### What it does
this function lets the user choose which cavern they want to enter 

#### Implementation (how it works)
The whileLoop method is responsible for handling the main game loop, where the player can choose a location to explore from a list of caverns. It repeatedly displays a list of cavern locations and prompts the player to make a choice. If the player selects a valid cavern, the game displays the chosen location and proceeds to the exploreCavern method, passing the relevant CavernRecord, the player's current points, and inventory. If the player chooses to exit, the game prints a farewell message and terminates the loop. The method ensures that the game continues until the player opts to exit


```java
public static void whileLoop(CavernArray caverns, int[] points, String[] inventory) {
    boolean exitGame = false;

    while (!exitGame) {
        System.out.println("Choose a location to explore:");
        for (int i = 0; i < caverns.records.length; i++) {
            System.out.println((i + 1) + ") " + caverns.records[i].location);
        }
        System.out.println((caverns.records.length + 1) + ") Exit");

        int choice = inputInt("Enter your choice:");

        if (choice >= 1 && choice <= caverns.records.length) {
            CavernRecord currentCavern = caverns.records[choice - 1];
            System.out.println("You chose " + currentCavern.location + "!");
            inventory = exploreCavern(currentCavern, points, inventory);
        } else if (choice == caverns.records.length + 1) {
            System.out.println("Exiting the game. Goodbye!");
            exitGame = true;
        } else {
            System.out.println("Invalid choice. Please choose a valid option.");
        }
    }
}
```

### exploreCavern

#### What it does
This function lets the player choose what they want to do in each location.

#### Implementation (how it works)
The exploreCavern method allows the player to explore a given cavern by providing three options: searching for objects, searching for monsters, or exiting the location. If the player chooses to search, they find an object in the cavern, gain points based on the object's value, and add the object to their inventory. The method then displays the updated inventory. If the player chooses to search for monsters, a monster appears, and the player engages in a fight. If the player opts to exit, the method ends the exploration of the cavern. 

```java
public static String[] exploreCavern(CavernRecord cavern, int[] points, String[] inventory) {
    boolean exitLocation = false;

    while (!exitLocation) {
        int choice = inputInt("Choose: 1) Search, 2) Search for Monsters, 3) Exit");
        if (choice == 1) {
            System.out.println("You find " + cavern.object + " worth " + cavern.objectsPoints + " points.");
            points[0] += cavern.objectsPoints;
            System.out.println("Your points are: " + points[0]);
            inventory = addToInventory(inventory, cavern.object);
            displayInventory(inventory, cavern);
        } else if (choice == 2) {
            System.out.println("You find a " + cavern.monster + " with " + cavern.monsterHealth + " health!");
            fightMonster(cavern, points, inventory);
        } else if (choice == 3) {
            System.out.println("Exiting " + cavern.location + "...");
            exitLocation = true;
        } else {
            System.out.println("Invalid choice. Please try again.");
        }
    }
    return inventory;
}
```

### fightMonster

#### What it does
this function gives the options from  the inventory to fight the monster

#### Implementation (how it works)
The fightMonster method allows the player to engage in combat with a monster found in a cavern. The player chooses a weapon from their inventory, and the damage dealt to the monster is calculated based on the weapon's damage points. The monster's health is reduced by the weapon's damage, and if the monster's health reaches zero, the player defeats it and earns points based on the monster's value. The method continues to prompt the player for a valid weapon choice until the monster is defeated. If the player chooses an invalid weapon, they are prompted again. The player's points are updated once the monster is vanquished.

```java
public static void fightMonster(CavernRecord cavern, int[] points, String[] inventory) {
    while (cavern.monsterHealth > 0) {
        System.out.println("The " + cavern.monster + " has " + cavern.monsterHealth + " health.");
        displayInventory(inventory, cavern);
        int weaponChoice = inputInt("Choose a weapon (1-" + inventory.length + "):");
        if (weaponChoice >= 1 && weaponChoice <= inventory.length) {
            System.out.println("You use " + inventory[weaponChoice - 1] + " dealing " + cavern.objectDamagePoints + " damage.");
            cavern.monsterHealth -= cavern.objectDamagePoints;
            if (cavern.monsterHealth > 0) {
                System.out.println("The monster still has " + cavern.monsterHealth + " health.");
            } else {
                System.out.println("You have defeated the " + cavern.monster + "!");
                points[0] += cavern.monsterPoints;
                System.out.println("Your points are now: " + points[0]);
            }
        } else {
            System.out.println("Invalid choice. Try again.");
        }
    }
}
```
